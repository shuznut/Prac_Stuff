```python
data['Height_m'] = data['Height']/100
```

```python
data['bmi']  = data['Weight']/(data['Height_m']**2)
```

```python
data.head(1)
```

```python
# create a list of conditions
conditions = [
    (data['bmi'] <= 18.5),
    (data['bmi'] > 18.5) & (data['bmi'] <= 24.9),
    (data['bmi'] > 25) & (data['bmi'] <=29.9 ),
    (data['bmi'] > 30)
    ]
```

```python
# create a list of the values we want to assign for each condition
values = ['Under_Weight', 'Normal', 'Over_weight', 'Obese']
```

```python
# create a new column and use np.select to assign values to it using our lists as arguments
data['category'] = np.select(conditions, values)
```

```python
data.head(2)
```

```python
sns.barplot(x="category", y="Gender", data=data, estimator=sum, palette="husl")
```

```python
condition_male = data['Gender']==1
condition_female = data['Gender']==0
condition_Normal = data['category']=='Normal' 
condition_Obese = data['category'] =='Obese'
condition_Under_Weight = data['category']=='Under_Weight'
condition_Over_Weight= data['category']=='Over_weight'

male_data1 = data[condition_male & condition_Normal ]
male_data2= data[condition_male & condition_Obese]
male_data3 = data[condition_male & condition_Under_Weight ]
male_data4 = data[condition_male & condition_Over_Weight]

female_data1 = data[condition_female & condition_Normal ]
female_data2= data[condition_female & condition_Obese]
female_data3 = data[condition_female & condition_Under_Weight ]
female_data4 = data[condition_female & condition_Over_Weight]

print('avg. normal weight amount of male   :',male_data1['bmi'].mean())
print('avg. Obese weight amount of male   :',male_data2['bmi'].mean())
print('avg. Under weight weight amount of male   :',male_data3['bmi'].mean())
print('avg. Over weight amount of male   :',male_data4['bmi'].mean())

print('avg. normal weight amount of female   :',female_data1['bmi'].mean())
print('avg. Obese weight amount of female   :',female_data2['bmi'].mean())
print('avg. Under weight weight amount of female   :',female_data3['bmi'].mean())
print('avg. Over weight amount of female   :',female_data4['bmi'].mean())
```
